```python
'''
Project 1: Student Management System
Objective
To manage student records using a console-based Python application.
Features
Add student details
View all students
Search student by roll number
Update marks
Delete student record
Save and load data from a file
Concepts Used
Variables, lists, dictionaries, loops, conditionals, functions, OOP, file handling, exception handling
Data Stored
Roll Number, Name, Marks
Sample Menu
1. Add Student
2. View Students
3. Search Student
4. Update Marks
5. Delete Student
6. Exit
'''

class StudentManagement:

    def __init__(self):
        self.students = []   # list to store student records

    def add_student(self):
        roll = int(input("Enter Roll Number: "))
        name = input("Enter Name: ")
        marks = int(input("Enter Marks: "))

        student = {
            "roll": roll,
            "name": name,
            "marks": marks
        }

        self.students.append(student)
        print("Student added successfully.")

    def view_students(self):
        if not self.students:
            print("No students found.")
            return

        print("\nRoll\tName\tMarks")
        for s in self.students:
            print(s["roll"], "\t", s["name"], "\t", s["marks"])

    def search_student(self):
        roll = int(input("Enter Roll Number to search: "))
        for s in self.students:
            if s["roll"] == roll:
                print("Student Found:", s)
                return
        print("Student not found.")

    def update_marks(self):
        roll = int(input("Enter Roll Number: "))
        for s in self.students:
            if s["roll"] == roll:
                s["marks"] = int(input("Enter new marks: "))
                print("Marks updated successfully.")
                return
        print("Student not found.")

    def delete_student(self):
        roll = int(input("Enter Roll Number to delete: "))
        for s in self.students:
            if s["roll"] == roll:
                self.students.remove(s)
                print("Student deleted successfully.")
                return
        print("Student not found.")


# Main Program
sms = StudentManagement()

while True:
    print("\n--- Student Management System ---")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Marks")
    print("5. Delete Student")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        sms.add_student()
    elif choice == "2":
        sms.view_students()
    elif choice == "3":
        sms.search_student()
    elif choice == "4":
        sms.update_marks()
    elif choice == "5":
        sms.delete_student()
    elif choice == "6":
        print("Program exited.")
        break
    else:
        print("Invalid choice.")
```

```python
'''Project 2: ATM / Bank Account Simulation
Objective
To simulate basic banking operations using Python.
Features
Create bank account
Deposit money
Withdraw money
Check balance
Save transaction history in a file
Concepts Used
Variables, operators, loops, conditionals, functions, OOP, file handling, exception handling
Sample Menu
1. Deposit
2. Withdraw
3. Check Balance
4. Transaction History
5. Exit
'''

class BankAccount:
    def __init__(self):
        self.balance = 0

    def deposit(self):
        amount = int(input("Enter amount to deposit: "))
        self.balance = self.balance + amount
        print("Deposit successful")

    def withdraw(self):
        amount = int(input("Enter amount to withdraw: "))
        if amount <= self.balance:
            self.balance = self.balance - amount
            print("Withdrawal successful")
        else:
            print("Insufficient balance")

    def check_balance(self):
        print("Current Balance:", self.balance)


# Main Program
account = BankAccount()

while True:
    print("\n--- ATM Menu ---")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        account.deposit()
    elif choice == "2":
        account.withdraw()
    elif choice == "3":
        account.check_balance()
    elif choice == "4":
        print("Thank you! Program exited.")
        break
    else:
        print("Invalid choice")
```

```python
'''
Project 3: Quiz Application (MCQ System)
Objective
To conduct a multiple-choice quiz using questions stored in a file.
Features
Load questions from file
Ask questions to user
Calculate score
Store final score in a file
Concepts Used
Lists, dictionaries, loops, conditionals, functions, file handling, exception handling
Sample Question Format
What is Python?
a) Snake
b) Programming Language
c) Game
Answer: b
'''

class Quiz:
    def __init__(self):
        self.score = 0

    def start_quiz(self):
        # Question 1
        print("1. What is Python?")
        print("a) Snake")
        print("b) Programming Language")
        print("c) Game")
        if input("Answer: ") == "b":
            self.score += 1

        # Question 2
        print("\n2. Which is a loop in Python?")
        print("a) if")
        print("b) while")
        print("c) break")
        if input("Answer: ") == "b":
            self.score += 1

        # Question 3
        print("\n3. Which symbol is used for comments?")
        print("a) //")
        print("b) #")
        print("c) /* */")
        if input("Answer: ") == "b":
            self.score += 1

        # Final Result
        print("\nQuiz Finished!")
        print("Your Score:", self.score)


# Main Program
quiz = Quiz()
quiz.start_quiz()
```

```python
'''
Project 4: Library Management System
Objective
To manage book records in a small library.
Features
Add new books
Issue books
Return books
View available books
Save records to a file
Concepts Used
Lists, dictionaries, loops, conditionals, functions, OOP, file handling, exception handling
Sample Menu
1. Add Book
2. Issue Book
3. Return Book
4. View Books
5. Exit
'''

class Library:
    def __init__(self):
        self.books = []

    def add_book(self):
        book = input("Enter book name: ")
        self.books.append(book)
        print("Book added.")

    def issue_book(self):
        book = input("Enter book name to issue: ")
        if book in self.books:
            self.books.remove(book)
            print("Book issued.")
        else:
            print("Book not available.")

    def return_book(self):
        book = input("Enter book name to return: ")
        self.books.append(book)
        print("Book returned.")

    def view_books(self):
        print("\nAvailable Books:")
        for book in self.books:
            print(book)


# Main Program
library = Library()

while True:
    print("\n--- Library Menu ---")
    print("1. Add Book")
    print("2. Issue Book")
    print("3. Return Book")
    print("4. View Books")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        library.add_book()
    elif choice == "2":
        library.issue_book()
    elif choice == "3":
        library.return_book()
    elif choice == "4":
        library.view_books()
    elif choice == "5":
        print("Program exited.")
        break
    else:
        print("Invalid choice.")
```

```python
'''
Project 5: Expense Tracker
Objective
To track daily expenses and calculate total spending.
Features
Add expenses with category
View expenses
Calculate total
Save expenses to file
Concepts Used
Variables, lists, dictionaries, loops, conditionals, functions, file handling, exception handling
Sample Expense Entry
Date: 05-12-2025
Category: Food
Amount: 250
'''

class ExpenseTracker:
    def __init__(self):
        self.expenses = []

    def add_expense(self):
        date = input("Enter date: ")
        category = input("Enter category: ")
        amount = int(input("Enter amount: "))

        self.expenses.append([date, category, amount])
        print("Expense added.")

    def view_expenses(self):
        print("\nDate\tCategory\tAmount")
        for e in self.expenses:
            print(e[0], "\t", e[1], "\t\t", e[2])

    def total_expense(self):
        total = 0
        for e in self.expenses:
            total = total + e[2]
        print("Total Expense:", total)



tracker = ExpenseTracker()

while True:
    print("\n--- Expense Tracker Menu ---")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Total Expense")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        tracker.add_expense()
    elif choice == "2":
        tracker.view_expenses()
    elif choice == "3":
        tracker.total_expense()
    elif choice == "4":
        print("Program exited.")
        break
    else:
        print("Invalid choice.")
```
