```python
# Mini Project: Student Result Management System
# using Core Python

# Taking input
name = input("Enter Student Name: ")

maths = int(input("Enter Maths Marks: "))
science = int(input("Enter Science Marks: "))
english = int(input("Enter English Marks: "))

# Calculating total
total = maths + science + english

# Calculating average
average = total / 3

# Grade calculation using if-else
if average >= 90:
    grade = "A+"
elif average >= 75:
    grade = "A"
elif average >= 60:
    grade = "B"
elif average >= 50:
    grade = "C"
elif average >= 35:
    grade = "Pass"
else:
    grade = "Fail"

# Final result check (subject-wise pass condition)
if maths < 35 or science < 35 or english < 35:
    result = "Fail"
else:
    result = "Pass"

# output
print("Result")
print("Student Name:", name)
print("Total Marks:", total)
print("Average Marks:", average)
print("Grade:", grade)
print("Final Result:", result)
```
